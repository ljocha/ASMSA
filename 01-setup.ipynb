## Define inputs and common parameters

```python
# Define input files
base = 'trpcage'

# input conformation, it should not contain hydrogens
conf = 'trpcage_correct.pdb'

# input trajectory
# atom numbering must be consistent with {conf}, no hydrogens as well

traj = 'trpcage_red.xtc'

# name the trained model in 04-train.ipynb, including file names used in later stages
model_name='unif'

# everything else is generated with pdb2gmx to make sure the files are consistent
```

```python
# likely unchanged -- when using sparse atom distances for features, specify its density (in how many pairs is an atom involved in average)
nb_density = 2 # integer in [1, n_atoms-1]
```

```python
topol = base + '.top'
index = base + '.ndx'
gro = base + '.gro'

# can specify other trajectory + pdb for more thorough statistics (05-plumed.ipynb)
full_conf = conf
full_xtc = traj

with open('inputs.py','w') as i:
    i.write(f'''
base = '{base}'
conf = '{conf}'
traj = '{traj}'
topol = '{topol}'
index = '{index}'
gro = '{gro}'
full_conf = '{full_conf}'
full_xtc = '{full_xtc}'
model_name = '{model_name}'

nb_density = {nb_density}
''')
```
